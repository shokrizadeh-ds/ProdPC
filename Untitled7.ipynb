```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

xx, yy = np.meshgrid(x, y, sparse=True)
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook

z = np.sin(xx ** 2 + yy ** 2)

plt.contour(x, y, z)
plt.show()
```

```python
plt.plot(x, y, z)
plt.show()
```
